### 미국의 중심 도시인 보스턴의 1970년대 거주지에 대한 데이터를 바탕으로 주택 가격을 예측해보자.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
data = pd.read_csv('data/boston_house.csv')
data
# 정답 컬럼은 MEDV이며 단위당 1000달
```

```python
data.info()
# 결측치 없음. 이상치 없음.
# 데이터타입은 문자열이 없음! -> 인코딩 불필요
# 각 컬럼마다 506개 데이터 있음.
```

```python
# 특성간 상관계수 출력
# 상관계수: 하나의 값이 변동될 때 나머지 하나의 값이 어떻게 변동되는지에 대한 계수(-1이면 완전반비례, 1이면 완전비례, 0이면 상관성이 없음.)
data.corr()
# corr: 기본적으로 피어슨 상관계수로 진행
```

```python
# 숫자로 보면 머리가 아프니까 히트맵으로 시각화를 해보자.
plt.figure(figsize=(15,15))
sns.heatmap(data = data.corr(),
            annot = True, # 각 셀 값 표시
            fmt = '.2f', # 소숫점 표시 자릿수
            annot_kws={'size':12}, # 수치 사이즈
            cmap='Blues') # 히트맵 색상(https://matplotlib.org/stable/tutorials/colors/colormaps.html)
```

### 주택가격(MEDV)과 상관관계가 높은 특성은
1. 양의 상관관계: RM(주택당 방 개수)
2. 음의 상관관계: LSTAT(소득 하위계층 비율)

### 모델링(모델 선택, 학습, 평가)
- 문제, 정답 및 학습, 평가 데이터로 분리시키기(학습, 평가 데이터 비율은 7:3)

```python
X = data.iloc[:,:-1]
y = data.loc[:,'MEDV']
```

```python
# train_test_split함수 쓰기!!!
X_train, X_test, y_train, y_test = train_test_split(X, # 문제
                                                    y, # 정답
                                                    test_size=0.3, # 평가 데이터의 비율(30%)
                                                    random_state=5) # 랜덤 시드값 설정
```

```python
print(X.shape)
print(y.shape)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### 선형회귀 모델 적용

```python
# 선형회귀 모델 임포트
from sklearn.linear_model import LinearRegression
# MSE 평가 지표 임포트
from sklearn.metrics import mean_squared_error
```

```python
# 선형회귀 모델 로드(모델 튜닝을 위한 하이퍼파라미터가 없음. 그래서 과대적합이 될 가능성이 큼.)
linear_model = LinearRegression()
```

```python
# 모델 학습
linear_model.fit(X_train, y_train)
```

```python
# 학습된 선형회귀 모델에 들어간 데이터의 특성 명칭 출력
# 13개가 나와야함. X_train의 컬럼 수가 13개라
linear_model.feature_names_in_
```

```python
# 각 특성별 weight(가중치)값 출력
# y = w1x1 + w2x2 + w3x3 + ...+w13x13 + b 의 w들만 출력
linear_model.coef_
```

```python
# 선형회귀 수식의 y절편값 출력
# y = w1x1 + w2x2 + w3x3 + ...+w13x13 + b 의 b를 출력
linear_model.intercept_
```

```python
# y = w1x1 + w2x2 + w3x3 + ...+w13x13 + b 가 만들어졌는데, 이 수식이 좋은 건지는 모르겠음. 평가를 해야함.
X
```

### 상관계수와 weight의 관계
- 정답 컬럼과 상관계수가 높을수록 해당 특성의 weight값의 절대값이 높아지는 경향이 있으나, weight값은 다른 특성들과의 조합에 따라 달라질 수 있으므로 상관계수가 높다고 해서 해당 특성의 weight가 반드시 높다고 볼 수 없음.
- 상관계수는 두 변수(특성)간의 상관성 정도이고, weight는 입력되는 모든 특성들을 다 고려하여 정답(y)을 예측함에 있어 중요한 정도라고 볼 수 있음.
- 따라서 weight값과 상관계수를 같이 고려해주는 것이 좋음.

### 모델 평가
- MSE, RMSE 확인

```python
# 학습된 선형회귀 모델로 정답 예측
pred_train = linear_model.predict(X_train)
pred_test = linear_model.predict(X_test)
```

```python
pred_train
```

```python
pred_test
```

```python
# MSE, RMSE 수치 확인
train_mse = mean_squared_error(pred_train, y_train)
test_mse = mean_squared_error(pred_test, y_test)

print('train MSE: ', train_mse)
print('test MSE: ', test_mse)
print('train MSE and test MSE gap: ', abs(train_mse - test_mse))

print('='*45)

# RMSE는 MSE의 루트값
print('train RMSE: ', train_mse**0.5) # **0.5 대신에 np.sqrt써도 무방
print('test RMSE: ', test_mse**0.5)
print('train RMSE and test RMSE gap: ', abs(train_mse**0.5 - test_mse**0.5))
```

```python
# 실제 정답(집 가격)의 평균 수치를 보자
print('train:', y_train.mean())
print('test:', y_test.mean())
```

- 실제 집가격의 평균을 보니 약 22,000만 달러인데, 전체오차(RMSE)가 4,300달러라면 상당히 큰 오차임.

## Scaling 적용
- 수치나 거리 기반 모델에는 스케일링을 적용하여 특성들 간 데이터 스케일을 맞춰주면 모든 특성에 대해 균형있게 weight에 영향을 미칠 수 있음.
- 스케일링이 없으면 특성간 범위 차이가 큰 경우 수치가 큰 특성의 weight가 커질 가능성이 있음.
- 스케일링한다고 해서 무조건 성능이 향상되는 건 아니라서 실제로 스케일링을 진행해서 효과가 있는지 테스트 해볼  필요가 있음.

```python
# 스케일러 객체 생성

scaler_SS = StandardScaler()
```

```python
# X_train데이터로 데이터의 범위들을 학습시킴
# test 데이터가 분리되어 있다면 항상 train데이터로만 학습시켜야함.
scaler_SS.fit(X_train) # 항상 train데이터로 학습시키기!!
```

```python
# 실제로 일정한 범위로 수치 변경
trans_X_train = scaler_SS.transform(X_train)
trans_X_test = scaler_SS.transform(X_test)
```

```python
X_train
```

```python
trans_X_train
# X_train은 컬럼마다 큰 수, 작은 수 뒤죽박죽인데, 작은 수들로 크기를 맞춤..
```

```python
# 스케일링이 된 데이터로 학습
linear_model2 = LinearRegression()
linear_model2.fit(trans_X_train, y_train)
```

```python
pred_train_SS = linear_model2.predict(trans_X_train)
pred_test_SS = linear_model2.predict(trans_X_test)
```

```python
# 스케일링 후 MSE, RMSE 수치 확인
train_mse = mean_squared_error(pred_train_SS, y_train)
test_mse = mean_squared_error(pred_test_SS, y_test)

print('train MSE: ', train_mse)
print('test MSE: ', test_mse)
print('train MSE and test MSE gap: ', abs(train_mse - test_mse))

print('='*45)

# RMSE는 MSE의 루트값
print('train RMSE: ', train_mse**0.5) # **0.5 대신에 np.sqrt써도 무방
print('test RMSE: ', test_mse**0.5)
print('train RMSE and test RMSE gap: ', abs(train_mse**0.5 - test_mse**0.5))
```

- 무조건 거리, 수치 기반 모델이라고 스케일링 시 효과가 있는 건 아니더라.

## 선형회귀 모델 특성확장
- 선형회귀 모델은 데이터의 분포가 선형 수식으로 표현하기 힘든 분포라면 예측 정확도가 많이 낮아질 수 있음.
- 데이터의 분포가 복잡해지면 회귀 직선만으로는 모든 데이터를 잘 설명할 수 없기 때문
- 선형회귀 모델을 모델 튜닝을 위한 하이퍼파라미터가 없음.
- 따라서 데이터의 분포 자체에 영향을 많이 받게 됨. 이 때 데이터의 복잡도를 높여서 모델이 패턴과 규칙을 더 잘 학습하게 만들어주면 성능을 높일 수 있음.
- 데이터의 복잡도를 높인다는 건 특성(컬럼)의 수를 늘려준다는 것

```python
# 특성들 간의 곱을 통해서 새로운 컬럼들을 만들어서 데이터의 복잡도를 높여보자
# DF를 만들기 위해 데이터를 넣을 딕셔너리 생성
extend_train_columns = {}

# X_train의 각 컬럼들을 서로서로 곱해서 만들어서 컬럼들을 딕셔너리에 추가
# 컬럼 13개 x 컬럼 13개해서 총 169개의 컬럼이 새로 만들어짐.
for col1 in X_train.columns:
    for col2 in X_train.columns:
        extend_train_columns[col1 + 'x' + col2] = X_train[col1] * X_train[col2]

extend_train_columns
```

```python
# 새롭게 만들어진 컬럼들을 데이터 프레임을 만들기
extend_train_df = pd.DataFrame(extend_train_columns)

# 기존 X_train에다가 concat으로 붙이기
# 13+169 = 182개의 컬럼이 나와야함.
extend_X_train = pd.concat([X_train, extend_train_df], axis=1) # axis=1: 왼쪽에서 오른쪽으로 붙여주기
extend_X_train
```

```python
# test데이터도 동일하게 특성확장 진행
extend_test_columns = {}

for col1 in X_test.columns:
    for col2 in X_test.columns:
        extend_test_columns[col1 + 'x' + col2] = X_test[col1] * X_test[col2]

extend_test_columns
```

```python
extend_test_df = pd.DataFrame(extend_test_columns)
extend_X_test = pd.concat([X_test, extend_test_df], axis=1) # axis=1: 왼쪽에서 오른쪽으로 붙여주기
extend_X_test
```

```python
# 특성확장된 train데이터의 문제와 정답으로 학습
linear_model3 = LinearRegression()
linear_model3.fit(extend_X_train, y_train)
```

```python
ex_pred_train = linear_model3.predict(extend_X_train)
ex_pred_test = linear_model3.predict(extend_X_test)
```

```python
# 스케일링 후 MSE, RMSE 수치 확인
ex_train_mse = mean_squared_error(ex_pred_train, y_train)
ex_test_mse = mean_squared_error(ex_pred_test, y_test)

print('train MSE: ', ex_train_mse)
print('test MSE: ', ex_test_mse)
print('train MSE and test MSE gap: ', abs(ex_train_mse - ex_test_mse))

print('='*45)

# RMSE는 MSE의 루트값
print('train RMSE: ', ex_train_mse**0.5) # **0.5 대신에 np.sqrt써도 무방
print('test RMSE: ', ex_test_mse**0.5)
print('train RMSE and test RMSE gap: ', abs(ex_train_mse**0.5 - ex_test_mse**0.5))
```

```python
linear_model3.score(extend_X_train, y_train)
```

```python
linear_model3.score(extend_X_test, y_test)
```

- 특성확장을 한 결과 절대적인 수치값이 올라가는 성능향상 효과가 있으나, train과 test의 차이가 큼...과대적합일 가능성이 있다.
- tran과 test의 편차를 줄일 수는 없을까
- 선형 모델의 과대적합을 줄여줄 수 있는 규제(Regulation)가 적용된 상위 모델들을 사용해보자.

## 라쏘(Lasso) 모델 적용
- L1 규제
- 경사하강법이 적용됨

```python
from sklearn.linear_model import Lasso
```

```python
lasso_model = Lasso(alpha = 10, # alpha는 규제의 강도로 높아지면 과대적합은 줄어들고 상대적으로 오차는 커질 수 있음.
                    max_iter = 10000) # max_iter: 최적 파라미터를 찾기 위한 경사하강법 반복 실행 횟수
```

```python
# 라쏘 모델 학습
lasso_model.fit(extend_X_train, y_train)
```

```python
# 라쏘는 L1규제로 중요도가 낮은 특성의 w값을 경사하강법을 통해 0으로 만들어줄 수 있기 때문에
# 0이 아닌 값들의 숫자를 세면 예측에 사용된 w값의 개수를 파악할 수 있음.
print(f'정답 예측에 사용한 특성 수: {sum(lasso_model.coef_ != 0)}')
```

```python
pred_train_la = lasso_model.predict(extend_X_train)
pred_test_la = lasso_model.predict(extend_X_test)
```

```python
train_mse_la = mean_squared_error(pred_train_la, y_train)
test_mse_la = mean_squared_error(pred_test_la, y_test)

print('train MSE: ', train_mse_la)
print('test MSE: ', test_mse_la)
print('train MSE and test MSE gap: ', abs(train_mse_la - test_mse_la))

print('='*45)

# RMSE는 MSE의 루트값
print('train RMSE: ', train_mse_la**0.5) # **0.5 대신에 np.sqrt써도 무방
print('test RMSE: ', test_mse_la**0.5)
print('train RMSE and test RMSE gap: ', abs(train_mse_la**0.5 - test_mse_la**0.5))
```

```python
lasso_model.score(extend_X_train, y_train)
```

```python
lasso_model.score(extend_X_test, y_test)
```

- 선형회귀에 비해 과대적합이 약간 해소됨.

## 릿지(Ridge) 모델 적용
- L2 규제

```python
from sklearn.linear_model import Ridge
```

```python
ridge_model = Ridge(alpha = 10,
                    max_iter=10000)
```

```python
ridge_model.fit(extend_X_train, y_train)
```

```python
pred_train_ridge = ridge_model.predict(extend_X_train)
pred_test_ridge = ridge_model.predict(extend_X_test)
```

```python
train_mse_ridge = mean_squared_error(pred_train_ridge, y_train)
test_mse_ridge = mean_squared_error(pred_test_ridge, y_test)

print('train MSE: ', train_mse_ridge)
print('test MSE: ', test_mse_ridge)
print('train MSE and test MSE gap: ', abs(train_mse_ridge - test_mse_ridge))

print('='*45)

# RMSE는 MSE의 루트값
print('train RMSE: ', train_mse_ridge**0.5) # **0.5 대신에 np.sqrt써도 무방
print('test RMSE: ', test_mse_ridge**0.5)
print('train RMSE and test RMSE gap: ', abs(train_mse_ridge**0.5 - test_mse_ridge**0.5))
```

```python
ridge_model.score(extend_X_train, y_train)
```

```python
ridge_model.score(extend_X_test, y_test)
```

- alpha값이 10일 때는 라쏘모델이 릿지모델에 비해 과대적합 해소에 유리함.

### 하이퍼파라미터 튜닝
- alpha값을 바꿔가며 RMSE의 변화를 확인해보기

```python
# lasso 모델
alpha_list = [0.001, 0.01, 0.1, 1, 10, 100, 1000]
lasso_train_list, lasso_test_list = [], []

for i in alpha_list:
    lasso_model = Lasso(alpha=i, max_iter=10000)
    lasso_model.fit(extend_X_train, y_train)

    # train
    lasso_train_pred = lasso_model.predict(extend_X_train)
    lasso_train_RMSE = mean_squared_error(lasso_train_pred, y_train)**0.5
    lasso_train_list.append(lasso_train_RMSE)

    # test
    lasso_test_pred = lasso_model.predict(extend_X_test)
    lasso_test_RMSE = mean_squared_error(lasso_test_pred, y_test)**0.5
    lasso_test_list.append(lasso_test_RMSE)
```

```python
plt.figure(figsize=(10,4))
plt.plot(lasso_train_list, label = 'train')
plt.plot(lasso_test_list, label = 'test')
plt.title('Lasso: alpha - RMSE chart')
plt.xlabel('alpha')
plt.xlabel('RMSE')
plt.xticks(np.arange(7), alpha_list)
plt.legend()
plt.grid()
```

```python
# ridge 모델
# lasso 모델
alpha_list = [0.001, 0.01, 0.1, 1, 10, 100, 1000]
ridge_train_list, ridge_test_list = [], []

for i in alpha_list:
    ridge_model = Ridge(alpha=i, max_iter=10000)
    ridge_model.fit(extend_X_train, y_train)

    # train
    ridge_train_pred = ridge_model.predict(extend_X_train)
    ridge_train_RMSE = mean_squared_error(ridge_train_pred, y_train)**0.5
    ridge_train_list.append(ridge_train_RMSE)

    # test
    ridge_test_pred = ridge_model.predict(extend_X_test)
    ridge_test_RMSE = mean_squared_error(ridge_test_pred, y_test)**0.5
    ridge_test_list.append(ridge_test_RMSE)
```

```python
plt.figure(figsize=(10,4))
plt.plot(ridge_train_list, label = 'train')
plt.plot(ridge_test_list, label = 'test')
plt.title('Ridge: alpha - RMSE chart')
plt.xlabel('alpha')
plt.xlabel('RMSE')
plt.xticks(np.arange(7), alpha_list)
plt.ylim(2, 6.5)
plt.legend()
plt.grid()
```

## 결론!
- 현재 보스턴 집값 데이터에서는 alpha가 10일 때, Ridge모델에 비해 Lasso 모델의 train, test 오차값의 차이가 적기 때문에 과대적합에 더 유리한 모델과 하이퍼파라미터라고 볼 수 있음.
