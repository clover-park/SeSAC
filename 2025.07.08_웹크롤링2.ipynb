### iframe 태그
- 다른 서버의 데이터나 리소스를 현재 서버의 웹 페이지에 포함시켜 보여주는 HTML 태그
- 웹 페이지에 보여줄 데이터가 많은 경우 한번에 접근하려면 그만큼 소모되는 자원이 많아 속도가 느려질 수 있는데, iframe을 사용하면 다른 서버에서 데이터를 불러와서 보여줄 수 있다.
- 기업에서 중요한 데이터를 따로 관리하기도 하는데 그럴 때 관리되는 서버에서 값을 빼와서 보여주는 용도로 사용함.
- 개발자도구를 봤는데 iframe 태그를 활용하여 데이터를 출력시켜주고 있다면 기존처럼의 접근은 불가하고 해당 데이터가 있는 원천 서버의 주소로 접근해야함.

```python
res_iframe = req.get('https://finance.naver.com/marketindex/exchangeList.naver', headers = U_A)
res_iframe
```

```python
soup_iframe = bs(res_iframe.text, 'lxml')
soup_iframe
```

```python
currency = soup_iframe.select('.tit > a')
currency
# 이제야 나온다!
```

```python
exchange_rate = soup_iframe.select('.sale')
exchange_rate
# 이제야 나온다!
```

```python
curr_list = []
rate_list = []

for i in range(len(currency)):
    curr_list.append(currency[i].text.strip())
    rate_list.append(exchange_rate[i].text)

rate_list
```

```python
naverpay_dic = {'currency':curr_list, 'exchange rate':rate_list}

naverpay_df = pd.DataFrame(naverpay_dic)
naverpay_df.index = range(1,len(curr_list)+1)
naverpay_df
```

```python
soup_review = bs(res_review.text, 'lxml')
soup_review
```

```python
review = soup_review.select('.review_list_v2__message.js-collapsed-review-content')
review
```

```python
# 리뷰별로 네이버 페이 구매평 이라는 텍스트가 있다면 지우고 새 리스트에 리뷰를 저장해보자.
review_list = []

for i in review:
    i = i.text.strip()
    if '네이버 페이 구매평' in i:
        new = i.replace(i[-38:],'')
        i = new
    review_list.append(i)

review_list
```

#### 1~10page의 리뷰를 모두 수집해보자.
- page가 달라지면 어떻게 접근해야할까: url에 page=2, page=3이런 거가 있다.

```python
# tqdm: 반복문의 진행상황을 %로 알려주는 모듈
# taqaddum이라는 아랍어라고 한다....뜻은 진행
from tqdm import tqdm
```

```python
review_all_list = []
page_list = []

# 1~10 page별 반복 코드
for i in tqdm(range(1,11)):
    # url 포매팅, page에만 {i}
    url = f"https://review4.cre.ma/bodyluv.kr/products/reviews?app=0&iframe=1&iframe_id=crema-product-reviews-3&install_method=hardcoded&page={i}&parent_url=https%3A%2F%2Fbodyluv.kr%2Fproduct%2F%25EB%25B0%2594%25EB%2594%2594%25EB%259F%25BD-%25EB%25B2%25A0%25EA%25B0%259C-best-6%25EC%25A2%2585%2F439%2Fcategory%2F65%2Fdisplay%2F1%2F&product_code=439&secure_device_token=V2cfca5208c34cdf3dce549fa4a0d6abed687f5aa1f433a66dc994d8081aa6dad4ebe8578141be79b4a618ae79520d9272&widget_env=100&widget_style="
    res = req.get(url,headers = U_A)
    soup = bs(res.text,'lxml')
    review = soup.select('div.review_list_v2__message.js-collapsed-review-content')

    for j in range(len(review)):
        body_re = review[j].text.strip()
        if '네이버 페이 구매평' in body_re:
            body_re = body_re.replace(body_re[-38:],'')
        review_all_list.append(body_re)
        page_list.append(f'Page{i}')
        
review_all_list
```

```python
review_dic = {'review':review_all_list, 'page':page_list}
review_df = pd.DataFrame(review_dic)
review_df.index = range(1,51)
pd.set_option('display.max_colwidth', None) # 내용 다 보이게
# 내용 좌측정렬
review_df.style.set_properties(**{"text-align":"left"}) # left, center, right 가능
```
